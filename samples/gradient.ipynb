```python
import numpy as np
from kdezero import Variable
from kdezero.utils import plot_dot_graph
import kdezero.functions as F
```

```python
x = Variable(np.array(1.0))
y = F.tanh(x)
x.name = 'x'
y.name = 'y'
```

```python
print(x, y)
```

```python
plot_dot_graph(y, verbose=False)
```

```python
y.backward(create_graph=True)
```

```python
# Calculate the first derivative
gx = x.grad
gx.name = 'gx1'
plot_dot_graph(gx, verbose=False)
```

```python
# Calculate the second derivative
x.cleargrad()
gx.backward(create_graph=True)

gx = x.grad
gx.name = 'gx2'
plot_dot_graph(gx, verbose=False)
```

```python
# Calculate the third derivative
x.cleargrad()
gx.backward(create_graph=True)

gx = x.grad
gx.name = 'gx3'
plot_dot_graph(gx, verbose=False)
```

```python
# Calculate the 4th derivative
x.cleargrad()
gx.backward(create_graph=True)

gx = x.grad
gx.name = 'gx4'
plot_dot_graph(gx, verbose=False)
```

```python
# Calculate the 5th derivative
x.cleargrad()
gx.backward(create_graph=True)

gx = x.grad
gx.name = 'gx5'
plot_dot_graph(gx, verbose=False)
```
